```
import  requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```
# converted the page into a text file and downloaded, then uploaded it into the session file and opened it directly

# Convert to a beautiful soup object
soup = bs(open('books-page.txt', 'rb').read(), features = 'html.parser')

# Print out our html
print(soup.prettify())
```

```
soup.find('body').prettify()
```

```
page1= soup.body.find('ol', class_= 'row')
```

```
print(page1.find('li').prettify())
```

```
book1 = page1.find('li')
cont= book1.select('div.image_container a img')[0]

bk_title = cont['alt']
bk_url = cont['src']

bk_price = book1.select('div.product_price p')[0].get_text()                                              
```

```
import re

bk_rating = book1.find('p', attrs= {'class': re.compile('star-rating')})['class'][1]
```

```
print(bk_title)
print(bk_url)
print(bk_price)
print(bk_rating)
```

```
table = []
pages = soup.find_all('ol', class_ = 'row')
for page in pages:
	books = page.find_all('li')
	for book in books:
		cont = book.select('div.image_container a img')[0]
		bk_title = cont['alt']

		bk_url = cont['src'].strip('./') # .strip() removes the unwanted characters in the link

		bk_price = book.select('div.product_price p')[0].get_text()
  
		bk_rating = book.find('p', attrs = {'class': re.compile('star-rating')})['class'][1]
    
		table.append([bk_title,bk_url,bk_price,bk_rating])
```

```
import pandas as pd

cols = ['Name', 'Image Url', 'Price', 'Rating']
df = pd.DataFrame(table, columns= cols)
df
```

```
rate= {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

df.Rating = [rate[item] for item in df.Rating] 
```
